```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
```

```python
#Processing for After High Max
```

```python
symbol = 'ACC'#this is kept for testing, it will not impact overral program
```

```python
#this function will return the combiled dataframe for After High Max
def build_df_comb_After_High_Max(df,symbol):
	#return df_filter_stock for After High Max scenario
	def build_df_filter_Stock_After_High_Max(df):
		#This is for After High Max
		cols_to_drop = ['Unnamed: 0','After_Low_Min']
		df_After_High_Max = df.drop(cols_to_drop, axis=1)
		#filter based on stock symbol
		df_filter_Stock_After_High_Max = df_After_High_Max.loc[df['Stock_Symbol']== symbol]
		return (df_filter_Stock_After_High_Max)
	
	df_filter_Stock_After_High_Max = build_df_filter_Stock_After_High_Max(df)
	#this function will return predicted value for After High Max
	def build_df_pred_After_High_Max(df_filter_Stock_After_High_Max):
		#function used for SVR model building for After High Max
		#build pipeline, standardize, fit the model and transform
		x = df_filter_Stock_After_High_Max.iloc[:, 2:9].values
		y = df_filter_Stock_After_High_Max.iloc[:,9:10].values
		x_train,x_test,y_train,y_test=train_test_split(x,y,test_size = 0.3, random_state = 1)
		pipeline = Pipeline([('scaler', StandardScaler()), ('estimator', SVR(kernel="rbf"))])
		y_scaler = StandardScaler()
		y_train = y_scaler.fit_transform(y_train)
		pipeline.fit(x_train, y_train)
		y_pred = y_scaler.inverse_transform(pipeline.predict(x_test))
		y_pred = np.round(y_pred,2)
		y_pred_lst = y_pred.tolist()#convert to list
		df_pred_After_High_Max = pd.DataFrame()#create a dataframe to store
		#df_pred['symbol'] = s
		df_pred_After_High_Max['y_pred_After_High_Max'] = y_pred_lst
		return df_pred_After_High_Max
	
	df_pred_After_High_Max = build_df_pred_After_High_Max(df_filter_Stock_After_High_Max)
	#This function will return combined dataframe of x_test and y_test
	def build_df_x_test_y_test_After_High_Max(df_filter_Stock_After_High_Max):
		#function to build df with the required columns for After High Max
		x = df_filter_Stock_After_High_Max.iloc[:, 0:9].values
		y = df_filter_Stock_After_High_Max.iloc[:,9:10].values
		x_train,x_test,y_train,y_test=train_test_split(x,y,test_size = 0.3, random_state = 1)
		x_test_lst = x_test.tolist()
		y_test_lst = y_test.tolist()
		df_x_test = pd.DataFrame()
		df_x_test = df_x_test.append(pd.DataFrame(x_test_lst),ignore_index=True)
		df_x_test.columns = ['Stock_Symbol','Result_Date','dt-7','dt-6','dt-5','dt-4','dt-3','dt-2','dt-1']
		#print("df_x_test is")
		#print(df_x_test)
		df_y_test = pd.DataFrame()
		df_y_test = df_y_test.append(pd.DataFrame(y_test_lst),ignore_index=True)
		df_y_test.columns = ['After_High_Max']
		#print("df_y_test is")
		#df_y_test
		df_x_test_y_test_After_High_Max = pd.concat([df_x_test, df_y_test], axis=1)
		#print("df_x_test_y_test_After_High_Max is")
		return df_x_test_y_test_After_High_Max
		
	df_x_test_y_test_After_High_Max = build_df_x_test_y_test_After_High_Max(df_filter_Stock_After_High_Max)
	
	df_comb_After_High_Max = pd.concat([df_x_test_y_test_After_High_Max, df_pred_After_High_Max], axis = 1)
    #print("df_comb_After_High_Max is")
    #print(df_comb_After_High_Max)
	return df_comb_After_High_Max
```

```python
df_comb_After_High_Max = build_df_comb_After_High_Max(df,symbol)
df_comb_After_High_Max
```

```python
#this function will return the combiled dataframe for After Low Min
def build_df_comb_After_Low_Min(df,symbol):
	#return df_filter_stock for After Low Min scenario
	def build_df_filter_Stock_After_Low_Min(df):
		#This is for After Low Min
		cols_to_drop = ['Unnamed: 0','After_High_Max']
		df_After_Low_Min = df.drop(cols_to_drop, axis=1)
		df_filter_Stock_After_Low_Min = df_After_Low_Min.loc[df['Stock_Symbol']== symbol]
		return (df_filter_Stock_After_Low_Min)

	df_filter_Stock_After_Low_Min = build_df_filter_Stock_After_Low_Min(df)
#this function will return predicted value for After Low Min
	def build_df_pred_After_Low_Min(df_filter_Stock_After_Low_Min):
		#function used for SVR model building for After Low Min
		#build pipeline, standardize, fit the model and transform
		x = df_filter_Stock_After_Low_Min.iloc[:, 2:9].values
		y = df_filter_Stock_After_Low_Min.iloc[:,9:10].values
		x_train,x_test,y_train,y_test=train_test_split(x,y,test_size = 0.3, random_state = 1)
		pipeline = Pipeline([('scaler', StandardScaler()), ('estimator', SVR(kernel="rbf"))])
		y_scaler = StandardScaler()
		y_train = y_scaler.fit_transform(y_train)
		pipeline.fit(x_train, y_train)
		y_pred = y_scaler.inverse_transform(pipeline.predict(x_test))
		y_pred = np.round(y_pred,2)
		y_pred_lst = y_pred.tolist()#convert to list
		df_pred_After_Low_Min = pd.DataFrame()#create a dataframe to store
		#df_pred['symbol'] = s
		df_pred_After_Low_Min['y_pred_After_Low_Min'] = y_pred_lst
		return df_pred_After_Low_Min

	df_pred_After_Low_Min = build_df_pred_After_Low_Min(df_filter_Stock_After_Low_Min)
#This function will return combined dataframe of x_test and y_test
	def build_df_x_test_y_test_After_Low_Min(df_filter_Stock_After_Low_Min):
	    #function to build df with the required columns for After Low Min
	    x = df_filter_Stock_After_Low_Min.iloc[:, 0:9].values
	    y = df_filter_Stock_After_Low_Min.iloc[:,9:10].values
	    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size = 0.3, random_state = 1)
	    x_test_lst = x_test.tolist()
	    y_test_lst = y_test.tolist()
	    df_x_test = pd.DataFrame()
	    df_x_test = df_x_test.append(pd.DataFrame(x_test_lst),ignore_index=True)
	    df_x_test.columns = ['Stock_Symbol','Result_Date','dt-7','dt-6','dt-5','dt-4','dt-3','dt-2','dt-1']
	    #print("df_x_test is")
	    #print(df_x_test)
	    df_y_test = pd.DataFrame()
	    df_y_test = df_y_test.append(pd.DataFrame(y_test_lst),ignore_index=True)
	    df_y_test.columns = ['After_Low_Min']
	    #print("df_y_test is")
	    #df_y_test
	    df_x_test_y_test_After_Low_Min = pd.concat([df_x_test, df_y_test], axis=1)
	    #print("df_x_test_y_test_After_High_Max is")
	    return df_x_test_y_test_After_Low_Min

	df_x_test_y_test_After_Low_Min = build_df_x_test_y_test_After_Low_Min(df_filter_Stock_After_Low_Min)

	df_comb_After_Low_Min = pd.concat([df_x_test_y_test_After_Low_Min, df_pred_After_Low_Min], axis = 1)
	return df_comb_After_Low_Min
```

```python
df_comb_After_Low_Min = build_df_comb_After_Low_Min(df,symbol)
df_comb_After_Low_Min
```

```python
#This function will combine the dataframe of After High Max and After Low Min
def build_df_combined_ind_stock(df_comb_After_Low_Min,df_comb_After_High_Max):
    df_comb_After_High_Max_mod = df_comb_After_High_Max[['After_High_Max','y_pred_After_High_Max']]
    df_combined_ind_stock = pd.concat([df_comb_After_Low_Min, df_comb_After_High_Max_mod], axis = 1)
    #print("df_combined_ind_stock is")
    #print(df_combined_ind_stock)
    return df_combined_ind_stock
```
